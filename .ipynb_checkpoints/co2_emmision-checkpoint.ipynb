### Step 2 : Split Data ( Train/Test ) 

```python
cdf = df[["Engine Size(L)","CO2 Emissions(g/km)"]]

msk = np.random.rand(len(cdf)) > 0.8
df_test = cdf[msk] 
df_train = cdf[~msk]
```

### Step 3 : Create Machine Learning Model

```python
# import libraries
from sklearn import linear_model
reg = linear_model.LinearRegression()

train_x = np.asanyarray(df_train[["Engine Size(L)"]])
train_y = np.asanyarray(df_train[["CO2 Emissions(g/km)"]])

reg.fit(train_x,train_y)

print("Coefficient : %0.2f" %reg.coef_[0][0])
print("Intercepyion : %0.2f" %reg.intercept_[0])
print("Model =>   Y = %0.2fX + %0.2f" %(reg.coef_[0][0],reg.intercept_[0])  )
```

```python
# plot model
fig = plt.figure()
ax = fig.add_subplot(111)
plt.ylabel("Engine Size")
plt.xlabel("Co2 Emissions")

ax.scatter(train_x,train_y)
ax.plot(train_x,train_x * reg.coef_[0][0] + reg.intercept_[0] , 'r--' )

plt.show()
```
